## Exploratory Analysis

Let's Understand our data first using pandas-profiling because it helps you to get intuition of your data structure in one line of code.

We can observe that the dataset has about 2717 duplicate values, we can drop those values to avoid overfitting when we fit the data

```python
#dropping the duplicate rows
new_df.drop_duplicates(inplace=True)
```

Let's start answering some questions with the data

### How often do people contribute to opensource

This can be seen from the exploratory analysis above that only 12.4% of developers contribute to open source often and 36.3% has never contributed before.

What type of employment do those that contribute OFTEN to opensourse often have. 

```python
new_df[new_df['OpenSourcer'] == 'Once a month or more often'].groupby('Employment').count()[['OpenSourcer']]
```

What type of employment do people that have contributed at least once in open source have

```python
new_df[new_df['OpenSourcer'] != 'Never' ].groupby('Employment').count()[['OpenSourcer']]
```

```python
new_df[new_df['OpenSourcer'] == 'Never' ].groupby('Employment').count()[['OpenSourcer']]
```

### Do they make more money than others who do not or is there a special satisfaction or rewards they get from doing so?

What is the average salary of those who contribute often to open source. From this data, this money includes both from  regular job, open source and others.

We need to convert all salary to weekly as some wrote their weekly salary, while some wrote their monthly and some yearly salary

```python
def conv(row):
    if row['CompFreq'] == 'Yearly':
        div = 52
    elif row['CompFreq'] == 'Monthly':
        div = 4
    else:
        div = 1
    return row['ConvertedComp'] / div

new_df['WeeklyPay'] = new_df.apply(conv, axis=1)
```

```python
new_df.head()
```

Now we are going to use the WeeklyPay column to find the average salary. First we will fill the missing values with median as there are outliers of some developers that make as much as 2 million and those that make $0.

```python
#filling the missing values with median
new_df['WeeklyPay'].fillna(new_df['WeeklyPay'].median(), inplace=True)
```

```python
#visualizing the data
plt.figure(figsize=(20,5))
sns.distplot(new_df['WeeklyPay'], kde=False, hist_kws={ 'alpha':1.0})
```

From the chart above it seems like 2000000 is an outlier but let's see the no of people make above 1 million

```python
np.sum(new_df['WeeklyPay'] == 2000000)
```

```python
# people that make less than the minimum wage using $7.25 per hour of average 42 week work hour
np.sum(new_df['WeeklyPay'] <= 305)
```

```python
new_df.groupby('OpenSourcer')[['WeeklyPay']].mean()
```

 We can see from above that contributing to open source does not necessarily mean more money.

### Years of Code Experience 

How many years of code experience does those who contribute often to open source have?

The YearsCode column is in object type so we need to change it to numerical value and fill in the missing values with mean as the missing values are about 1%

```python
#changing More than 50 years to 50 as converting to float will not work with strings
new_df['YearsCode'] = np.where(new_df.YearsCode == 'More than 50 years', '50', new_df.YearsCode)
```

```python
#changing Less than 1 year to 1
new_df['YearsCode'] = np.where(new_df.YearsCode == 'Less than 1 year', '1', new_df.YearsCode)
```

```python
#converting the column from string to float
new_df['YearsCode'] = new_df['YearsCode'].astype('float')
```

```python
#filling the missing values with mean
new_df['YearsCode'].fillna(new_df['YearsCode'].mean(), inplace=True)
```

```python
new_df.groupby('OpenSourcer')[['YearsCode']].mean()
```

The above chart shows that years of experience does not matter when contributing to open source. From below about 2.3% of people with less that 5 years of code experience are active contributors

```python
new_df[new_df['YearsCode'] < 5 ].groupby('OpenSourcer').count()[['YearsCode']]
```
